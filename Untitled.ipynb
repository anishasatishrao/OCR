```python
import easyocr
import cv2
import matplotlib.pyplot as plt

image_path = "C:/Users/Anisha.S/Documents/30days Projects/rc.jpg"
img = cv2.imread(image_path)

reader = easyocr.Reader(['en'], gpu=False)
text_ = reader.readtext(img)

for t in text_:
    bbox, text, score = t
    # Draw rectangle around the detected text bbox[]
    cv2.rectangle(img, tuple(map(int, bbox[0])), tuple(map(int, bbox[2])), (0, 255, 0), 2)
    # Put the detected text
    cv2.putText(img, text, tuple(map(int, bbox[1])), cv2.FONT_HERSHEY_COMPLEX, 0.65, (255, 0, 0), 2)

# Display the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()


# import cv2
# import easyocr
# import matplotlib.pyplot as plt

# # 1. Open the webcam (0 = default camera)
# cap = cv2.VideoCapture(0)

# print("Press 's' to capture image, 'q' to quit")

# while True:
#     ret, frame = cap.read()
#     if not ret:
#         print("Failed to grab frame")
#         break

#     cv2.imshow("Camera - Press 's' to scan", frame)

#     key = cv2.waitKey(1)

#     if key == ord('s'):
#         # Save the captured frame
#         img = frame.copy()
#         break
#     elif key == ord('q'):
#         cap.release()
#         cv2.destroyAllWindows()
#         exit()

# cap.release()
# cv2.destroyAllWindows()

# # 2. Initialize OCR reader for Kannada
# reader = easyocr.Reader(['kn'], gpu=False)

# # 3. Run OCR
# text_ = reader.readtext(img)

# # 4. Draw results on image
# for bbox, text, score in text_:
#     cv2.rectangle(img, tuple(map(int, bbox[0])), tuple(map(int, bbox[2])), (0, 255, 0), 2)
#     cv2.putText(img, text, tuple(map(int, bbox[0])), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255, 0, 0), 2)

# # 5. Display result
# plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
# plt.title("Detected Kannada Text")
# plt.axis('off')
# plt.show()
```
